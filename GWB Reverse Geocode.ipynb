```python
table = pd.read_csv('GWB_RDMSwithLATLON.csv')
table
```

```python
geocodeAddress = []
```

```python
for index, row in table.iterrows():
    results = reverse_geocode([row['LON'], row['LAT']])
    geocodeAddress.insert(index, results['address']['LongLabel'])
    print(index)
```

```python
table['EsriAddress'] = geocodeAddress
```
